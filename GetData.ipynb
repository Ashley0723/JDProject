```python
# Set the header file information in the request
headers = {'User-Agent':'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Mobile Safari/537.36'}
# Set Address
url0 = 'https://club.jd.com/comment/productPageComments.action?callback=fetchJSON_comment98&pageSize=10&productId=100039924870&score=3&sortType=5&page='
```

```python
# Send a request and loop to crawl the page
for i in range(200):
    a = 0
    k = str(i)
    if i == a:
        url=(url0+k)
        r=requests.get(url=url,headers=headers)
        html=r.content
    else:
        k=str(i)
        url=(url0+k)
        r=requests.get(url=url,headers=headers)
        html2=r.content
        html = html + html2
    if i % 100 == 0:
        print("当前抓取页面:",url,"状态:",r)
```

```python
# Encode the crawled pages
html = str(html, encoding = "utf-8")

# Output the encoded page as txt text for storage
file = open(r'E:\Graduate\2024.09\OPEN-SOURCE-TOOL-FOR-DATA-SCIENCE\final\jd_comments_pos.txt', 'w', encoding = "utf-8")
file.write(html)
file.close()
```

```python
# Extract nickname
nickname = re.findall(r'"usefulVoteCount".*?,"nickname":(.*?),',html)

# Extract rating
score = re.findall(r'"creationTime".*?,"score":(.*?),',html)

# Extract the type of computer
productColor = re.findall(r'"creationTime".*?,"productColor":(.*?),',html)

# Extract evaluation interval days
days = re.findall(r'"usefulVoteCount".*?,"days":(.*?)}',html)
```

```python
# 使用正则提取评论信息
content1 = re.findall(r'"guid".*?,"content":(.*?),',html)
# 对提取的评论信息进行去重
content = []
for i in content1:
    if not "img" in i:
        content.append(i)

# 使用正则提取时间字段信息
creationTime1 = re.findall(r'"creationTime":(.*?),"referenceName',html)
# 提取日期和时间
creationTime = []
for d in creationTime1:
    date = d[1:20]
    creationTime.append(date)

#提取小时信息
hour = []
for h in creationTime:
    date = h[10:13]
    hour.append(date)
```

```python
all_data = pd.DataFrame({'creationTime':creationTime,'hour':hour,'nickname':nickname,'productType':productColor,'score':score,'content':content,'days':days})

#查看整理完的数据表
all_data
```
